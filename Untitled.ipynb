some visualization of the data

```r
options(repr.plot.width = 20, repr.plot.height = 10)

firstage<-selected|>
    filter(age=="18–29")
graph2 <- ggplot(firstage, aes(x=week_end, y=numtotal_atleast1dose))+
    geom_point()+
    xlab("Date") +
    ylab("Number of New Vaccinations") +
    theme(text = element_text(size = 12))        

graph2
```

```r
options(repr.plot.width = 10, repr.plot.height = 50)

graph1 <- ggplot(selected, aes(x=week_end, y=numtotal_atleast1dose, color=age))+
    geom_point()+
    facet_grid(rows = vars(age))+
    xlab("Date") +
    ylab("Number of New Vaccinations") +
    labs(color="Age Cathegory")+
    theme(text = element_text(size = 12))        

graph1
```
